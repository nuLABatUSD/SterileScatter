```python
# Imports
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.colors as colors

# Constants
_PI_ = 3.141592653589793
_electron_mass_ = 0.511
_muon_mass_ = 105.658
_neutral_pion_mass_ = 134.9768
_charged_pion_mass_ = 139.57039

# Initialize arrays
m_s = np.linspace(150,500,351)
E0_1 = np.zeros(len(m_s))
E0_2 = np.zeros(len(m_s))
Emin_3 = np.zeros(len(m_s))
Emax_3 = np.zeros(len(m_s))
Emin_4 = np.zeros(len(m_s))
Emax_4 = np.zeros(len(m_s))

# E0_1
E0_1 = m_s/2

# E0_2
E0_2 = (m_s**2 - _neutral_pion_mass_**2)/(2*m_s)

# Emin_3 
E_pi_3 = (m_s**2 + _charged_pion_mass_**2 - _electron_mass_**2)/(2*m_s) 
gamma_pi_3 = E_pi_3 / _charged_pion_mass_

p_pi_3 = np.sqrt(E_pi_3**2 - _charged_pion_mass_**2)
v_pi_3 = p_pi_3 / E_pi_3

E_nu_prime = (_charged_pion_mass_**2 - _muon_mass_**2)/(2*_charged_pion_mass_)
p_nu_prime = E_nu_prime

Emin_3 = gamma_pi_3 * (E_nu_prime - v_pi_3 * p_nu_prime)

# Emax_3
Emax_3 = gamma_pi_3 * (E_nu_prime + v_pi_3 * p_nu_prime)

# Emin_4
E_pi_4 = (m_s**2 + _charged_pion_mass_**2 - _muon_mass_**2)/(2*m_s)
gamma_pi_4 = E_pi_4 / _charged_pion_mass_

p_pi_4 = np.sqrt(E_pi_4**2 - _charged_pion_mass_**2)
v_pi_4 = p_pi_4 / E_pi_4

Emin_4 = gamma_pi_4 * (E_nu_prime - v_pi_4 * p_nu_prime)

# Emax_4
Emax_4 = gamma_pi_4 * (E_nu_prime + v_pi_4 * p_nu_prime)

# Plot
fig, ax = plt.subplots()
ax.plot(m_s,E0_1,label='$E_{01}$')
ax.plot(m_s,E0_2,label='$E_{02}$')
ax.plot(m_s,Emin_3,label='$E_{min3}$')
ax.plot(m_s,Emax_3,label='$E_{max3}$')
ax.plot(m_s,Emin_4,label='$E_{min4}$')
ax.plot(m_s,Emax_4,label='$E_{max4}$')
plt.ylabel('E [MeV]')
plt.xlabel('$m_s$ [MeV]')
plt.legend()
plt.show()


```
