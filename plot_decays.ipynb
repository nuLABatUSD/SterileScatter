```python
"""
Title: plot_decays.ipynb
Description: Plots dP/dtdE vs. E (we are calling this the Neutrino Decay Spectrum)
    - Plots for the 6 different flavors plus a total
    - Designed to read in the output files from decays_energy.cc
    
- Can adjust output_num and state_index to select file and state 
- output_num (int): number of the sim_output file
- state_index (int): index of the state to plot
"""
# Imports
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.colors as colors
import csv

folder_label = "decaysANDenergy"
output_num = 1;
#state_index = -1;

Neff = []
tau = []

data = np.loadtxt(folder_label+"/decays_output.csv", delimiter=',')
data2 = np.loadtxt(folder_label+"/eps_output.csv", delimiter=',')

energy = data2[0,:]

fig, ax = plt.subplots()

total_decays = np.zeros(len(energy))

for i in range(6):
    #ax.plot(energy,data[:,i],label=f"{i}", marker='o', markersize=2)
    ax.plot(energy,data[:,i],label=f"{i}")

    total_decays += data[:,i]
ax.plot(energy,total_decays,label="total")
#ax.plot(energy,total_decays,label="total",marker='o', markersize=2)
#plt.title(f"sim_output{i+1}.csv" + " $a_0$=" + str(a[0]) + "--> $a_f$=" + str(round(a[-1],3)))
plt.xlabel('$E_\\nu$')
plt.ylabel(r'$\dfrac{dP}{dtdE}$')
plt.title("Neutrino Decay Spectrum")
plt.legend()
plt.show()
```
