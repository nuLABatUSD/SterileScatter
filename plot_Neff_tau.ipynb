```python
"""
Title: plot_Neff_tau.ipynb
Description: Plots Neff vs. tau (lifetime)

The output folders need to have a naming scheme
For instance: test_1, test_2, test_3, ...
- folder_label (string): folder label
- num_runs (int): number of runs
"""
# Imports
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.colors as colors
import csv

# Adjustable Input
folder_label = 'test_'
num_runs = 2

Neff = []
tau = []

for i in range(num_runs):
    data = np.loadtxt(folder_label+f"{i+1}/Neff_output.csv", delimiter=',')
    Neff.append(data[0])
    tau.append(data[2])

fig, ax = plt.subplots()
ax.plot(tau,Neff,label='total', marker='o', markersize=2)

plt.title("$N_{eff}$ vs. $\\tau$")
plt.xlabel('$\\tau$')
plt.ylabel('$N_{eff}$')
plt.legend()
plt.show()
```
