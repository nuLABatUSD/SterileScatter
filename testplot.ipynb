```python
# Imports
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.colors as colors
import csv

file_values = np.zeros((8,4))

with open('output2.csv', mode ='r') as file:
    csvFile = csv.reader(file)
    i = 0
    for lines in csvFile:
        file_values[i] = lines
        i+=1
np.transpose(file_values,(1,0))

a = file_values[0]
T = file_values[2]

T_cm = 1/a
print(a)

#g_star = np.zeros(len(rho))

#g_star = rho/(T**4)/((np.pi**2)/30)

fig, ax = plt.subplots()
ax.plot(T_cm, T)
ax.semilogx(T_cm, T)
ax.invert_xaxis() # Invert the x axis
plt.show()
```

```python
# Imports
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.colors as colors
import csv

a = np.zeros(8)
T = np.zeros(8)

with open('output2.csv', mode ='r') as file:
    csvFile = csv.reader(file)
    i = 0
    for lines in csvFile:
        a[i] = (lines[0])
        T[i] = (lines[2])
        i+=1


T_cm = 1/a

T_1 = T_cm

fig, ax = plt.subplots()
ax.plot(T_cm, T)
ax.plot(T_cm, T_1)
plt.xscale('log')
plt.yscale('log')
ax.invert_xaxis() # Invert the x axis
plt.show()
```
