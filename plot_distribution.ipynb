```python
# Imports
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.colors as colors
import csv

for i in range(5):
    data = np.loadtxt(f"outputs/sim_output{i+1}.csv", delimiter=',')
    data2 = np.loadtxt(f"outputs/eps_output{i+1}.csv", delimiter=',')
    
    a = data[:,0]
    da = data[:,1]
    T = data[:,-1]
    t = data[:,-2]
    d = data[:,-3]
    
    eps = data2[0,:]
    
    f_i = data[0,2:2+200]
    f_f = data[-1,2:2+200]
    
    initial = eps**3 * f_i
    final = eps**3 * f_f
    
    
    fig, ax = plt.subplots()
    #ax.loglog(eps,initial,markersize = 3,marker='o',label='initial')
    ax.loglog(eps,initial,label='initial')
    ax.loglog(eps, final,label='final')
    plt.title(f"sim_output{i+1}.csv" + " $a_0$=" + str(a[0]) + "--> $a_f$=" + str(round(a[-1],3)))
    plt.xlabel('$\\epsilon$')
    plt.ylabel('$\\epsilon^3$f')
    plt.legend()
plt.show()
```
