```python
"""
Title: plot_distributions6.ipynb
Description: Print the spectra from all 6 neutrino types for a given file

Can adjust variables to select file and state 
- folder_name (string): name of folder
- output_num (int): number of the sim_output file
- state_index (int): index of the state to plot

"""
# Imports
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.colors as colors
import csv

# Adjustable Variables
folder_name = 'test1'
output_num = 81;
state_index = -1;

# Read In Data
data = np.loadtxt(folder_name+f"/sim_output{output_num}.csv", delimiter=',')
data2 = np.loadtxt(folder_name+f"/eps_output.csv", delimiter=',')

eps = data2[output_num*2 -2,:]
total = np.zeros(len(eps))

# Plot
fig, ax = plt.subplots()
for i in range(6):
    dist = data[state_index,2+(200*i):2+(200*(i+1))]
    
    spectrum = eps**3 * dist
    total += spectrum

    #ax.semilogy(eps,spectrum,label=f'{i}')
    ax.plot(eps,spectrum,label=f'{i}')

#ax.semilogy(eps,total,label='total',markersize = 3,marker='o')
ax.plot(eps,total,label='total')
#plt.title(f"sim_output{i+1}.csv" + " $a_0$=" + str(a[0]) + "--> $a_f$=" + str(round(a[-1],3)))
plt.xlabel('$\\epsilon$')
plt.ylabel('$\\epsilon^3$f')
#plt.xlim(0,60)
#plt.ylim(10e-10,10e2)
plt.legend()
plt.show()
```
